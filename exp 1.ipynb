```python
from google.colab import files
import pandas as pd

# Upload Spotify_Youtube.csv
uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

print("Dataset Loaded Successfully!")
print(df.head())
print(df.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# ---------------------------------
# Phase 1: Business Understanding
# Objective: Predict whether a song is Licensed
# ---------------------------------

# ---------------------------------
# Phase 2: Data Understanding
# ---------------------------------
crisp_df = df.copy()

# ---------------------------------
# Phase 3: Data Preparation
# ---------------------------------

drop_cols = [
    "Artist",
    "Track",
    "Album",
    "Title",
    "Channel",
    "Description",
    "Url_spotify",
    "Url_youtube",
    "Uri"
]

for col in drop_cols:
    if col in crisp_df.columns:
        crisp_df.drop(col, axis=1, inplace=True)

le = LabelEncoder()

for col in crisp_df.select_dtypes(include=['object']).columns:
    crisp_df[col] = le.fit_transform(crisp_df[col].astype(str))

for col in crisp_df.select_dtypes(include=['bool']).columns:
    crisp_df[col] = crisp_df[col].astype(int)

X = crisp_df.drop("Licensed", axis=1)
y = crisp_df["Licensed"]

imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# ---------------------------------
# Phase 4: Modeling
# ---------------------------------
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# ---------------------------------
# Phase 5: Evaluation
# ---------------------------------
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("CRISP-DM Accuracy:", accuracy)

# ---------------------------------
# Phase 6: Deployment (Optional)
# ---------------------------------
# import joblib
# joblib.dump(model, "spotify_model.pkl")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# -----------------------------
# Step 1: Selection
# -----------------------------
kdd_df = df.copy()

drop_cols = [
    "Artist",
    "Track",
    "Album",
    "Title",
    "Channel",
    "Description",
    "Url_spotify",
    "Url_youtube",
    "Uri"
]

for col in drop_cols:
    if col in kdd_df.columns:
        kdd_df.drop(col, axis=1, inplace=True)

# -----------------------------
# Step 2: Preprocessing
# -----------------------------
le = LabelEncoder()

for col in kdd_df.select_dtypes(include=['object']).columns:
    kdd_df[col] = le.fit_transform(kdd_df[col].astype(str))

for col in kdd_df.select_dtypes(include=['bool']).columns:
    kdd_df[col] = kdd_df[col].astype(int)

X = kdd_df.drop("Licensed", axis=1)
y = kdd_df["Licensed"]

imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)

# -----------------------------
# Step 3: Transformation
# -----------------------------
scaler = StandardScaler()
X = scaler.fit_transform(X)

# -----------------------------
# Step 4: Data Mining
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# -----------------------------
# Step 5: Interpretation
# -----------------------------
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("KDD Accuracy:", accuracy)
```
