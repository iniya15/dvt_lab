```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("/Users/student/Downloads/Spotify_Youtube.csv")

# Display first 5 rows
df.head()


# ---------------------------------------------------
# 1. Top 10 Artists by YouTube Views
# ---------------------------------------------------

top_artists_views = (
    df.groupby("Artist")["Views"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

print("Top 10 Artists by YouTube Views")
display(top_artists_views)

plt.figure(figsize=(10,5))

plt.bar(
    top_artists_views["Artist"],
    top_artists_views["Views"]
)

plt.title("Top 10 Artists by YouTube Views")
plt.xlabel("Artist")
plt.ylabel("Total Views")
plt.xticks(rotation=45)

plt.show()



# ---------------------------------------------------
# 2. Top 10 Artists by Spotify Streams
# ---------------------------------------------------

top_artists_streams = (
    df.groupby("Artist")["Stream"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

print("Top 10 Artists by Spotify Streams")
display(top_artists_streams)

plt.figure(figsize=(10,5))

plt.bar(
    top_artists_streams["Artist"],
    top_artists_streams["Stream"]
)

plt.title("Top 10 Artists by Spotify Streams")
plt.xlabel("Artist")
plt.ylabel("Total Streams")
plt.xticks(rotation=45)

plt.show()



# ---------------------------------------------------
# 3. Top 10 Songs by YouTube Views
# ---------------------------------------------------

top_songs_views = (
    df.groupby("Track")["Views"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

print("Top 10 Songs by YouTube Views")
display(top_songs_views)

plt.figure(figsize=(12,5))

plt.bar(
    top_songs_views["Track"],
    top_songs_views["Views"]
)

plt.title("Top 10 Songs by YouTube Views")
plt.xlabel("Song")
plt.ylabel("Total Views")
plt.xticks(rotation=90)

plt.show()



# ---------------------------------------------------
# 4. Album Type Distribution
# ---------------------------------------------------

album_type = (
    df["Album_type"]
    .value_counts()
    .reset_index()
)

album_type.columns = ["Album Type", "Count"]

print("Album Type Distribution")
display(album_type)

plt.figure(figsize=(8,5))

plt.bar(
    album_type["Album Type"],
    album_type["Count"]
)

plt.title("Album Type Distribution")
plt.xlabel("Album Type")
plt.ylabel("Number of Songs")

plt.show()
```
